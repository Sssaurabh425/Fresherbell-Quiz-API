```python
import re
```

```python
r = re.compile(r'<[^>]+>')
def remove_html(string):
    return r.sub('',string)
```

```python
quizzes['question'] = quizzes['question'].transform(remove_html)
```

```python
quizzes.isnull().sum()
```

```python
quizzes = quizzes.fillna("")
```

```python
quizzes['option1'] = quizzes['option1'].transform(remove_html)
quizzes['option2'] = quizzes['option2'].transform(remove_html)
quizzes['option3'] = quizzes['option3'].transform(remove_html)
quizzes['option4'] = quizzes['option4'].transform(remove_html)
quizzes['option5'] = quizzes['option5'].transform(remove_html)
```

```python
quizzes.duplicated().sum()
```

```python
quizzes['tags'] = quizzes['option1'] +" "+ quizzes['option2'] +" "+  quizzes['option3'] +" "+  quizzes['option4'] +" "+  quizzes['option5']
```

```python
quizzes = quizzes[['id','question','tags','subtopic_id']]
```

```python
quizzes.head()
```

```python
quizzes['question'] = quizzes['question'].transform(lambda x:x.lower())
```

```python
quizzes['tags'] = quizzes['tags'].transform(lambda x:x.lower())
```

```python
quizzes['tags'] = quizzes['question'] +" "+ quizzes['tags']
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
    return " ".join(y)
```

```python
stem('dance')
```

```python
quizzes['tags'] = quizzes['tags'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000, stop_words='english')
```

```python
vectors = cv.fit_transform(quizzes['tags']).toarray()
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity = cosine_similarity(vectors)
```

```python
sorted(list(enumerate(similarity[300])), reverse=True, key = lambda x:x[1]+1)[1:6]
```

```python
def recommend(quiz):
    q_index = quizzes[quizzes['question'] == quiz].index[0]
    distances = similarity[q_index]
    quiz_list = sorted(list(enumerate(distances)), reverse=True, key = lambda x:x[1])[1:6]
    
    for i in quiz_list:
        print(quizzes.iloc[i[0]].question)
```

```python
recommend(quizzes['question'][315])
```
